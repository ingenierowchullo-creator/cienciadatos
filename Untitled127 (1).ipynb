```python
# =========================================
# IMPORTAR LIBRERÍAS
# =========================================
import pandas as pd
from sklearn.linear_model import LogisticRegression
# =========================================
# CREAR DATOS
# =========================================
datos = {
    'horas_estudio': [1,2,3,4,5,6,7,8],
    'asistencia': [50,55,60,65,70,80,90,95],
    'aprueba': [0,0,0,0,1,1,1,1]
}
df = pd.DataFrame(datos)
print(df)
# =========================================
# VARIABLES
# =========================================
X = df[['horas_estudio', 'asistencia']]
y = df['aprueba']
# =========================================
# CREAR MODELO
# =========================================
modelo = LogisticRegression()
# =========================================
# ENTRENAR MODELO
# =========================================
modelo.fit(X, y)
# =========================================
# PREDICCIÓN
# =========================================
nuevo_estudiante = [[5, 75]]
prediccion = modelo.predict(nuevo_estudiante)
# =========================================
# RESULTADO
# =========================================
if prediccion[0] == 1:
    print("El estudiante APROBARÁ")
else:
    print("El estudiante NO aprobará")
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# DATOS
datos = {
    'edad': [18,22,25,30,35,40,45,50],
    'ingreso': [800,1000,1200,1500,1800,2200,2500,3000],
    'compra': [0,0,0,1,1,1,1,1]
}

df = pd.DataFrame(datos)

# VARIABLES
X = df[['edad', 'ingreso']]
y = df['compra']

# MODELO
modelo = LogisticRegression()

# ENTRENAMIENTO
modelo.fit(X, y)

# PREDICCIÓN
nuevo_cliente = [[28, 1600]]

resultado = modelo.predict(nuevo_cliente)

# RESULTADO
if resultado[0] == 1:
    print("El cliente COMPRARÁ")
else:
    print("El cliente NO comprará")
```
